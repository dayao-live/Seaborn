```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
crash_df=sns.load_dataset('car_crashes')
crash_df
```

# Distribution Plots


# Distribution plot

```python
sns.distplot(crash_df['not_distracted'],bins=25)
```

# Joint Plot

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
```

# KDE plot

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='kde') #kde:kernal density estimation
```

```python
sns.kdeplot(crash_df['alcohol'])
```

# Pair Plots

```python
sns.pairplot(crash_df)
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df,hue='sex',palette='Blues')
```

# Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

# Styling

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,4)) 
sns.set_context('paper',font_scale=1.4) #talk, poster,paper
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
sns.despine(left=True)
```

# Categorical Plots


# Bar Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips_df,estimator=np.median) #estimator could be np.var, np.mean, np.std, etc
```

# Count Plot

```python
sns.countplot(x='sex',data=tips_df)
```

# Box Plot

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips_df)
plt.legend(loc=1)
```

# Violin plot

```python
sns.violinplot(data=tips_df, x='day',y='total_bill',hue='sex',split=True) # a combination of KDE plot and boxplot
```

# Strip Plot

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day',y='total_bill',data=tips_df,jitter=True,hue='sex',dodge=True)
```

# Swarm Plot

```python

sns.swarmplot(x='day',y='total_bill',data=tips_df)
```

# Palettes

```python
plt.figure(figsize=(8,5))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(data=tips_df,x='day',y='total_bill',hue='sex',palette='afmhot') #matplotlib colormap
plt.legend(loc=4)
```

# Matrix Plot

# heatmaps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
crash_mx=crash_df.corr() #correlation matrix
crash_mx
sns.heatmap(crash_mx,annot=True,cmap='Blues')
```

```python
flights=sns.load_dataset('flights')
flights
flights=flights.pivot_table(index='month',columns='year',values='passengers')
flights
sns.heatmap(flights,cmap='Blues',linecolor='white',linewidth=1)
```

# PairGrid

```python
iris=sns.load_dataset('iris')
iris_g=sns.PairGrid(iris,hue='species')
#iris_g.map(plt.scatter)
```

```python
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris_g=sns.PairGrid(iris,hue='species',x_vars=['sepal_length','sepal_width'],y_vars=['petal_length','petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```
